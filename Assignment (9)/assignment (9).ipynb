# Dimensionality Reduction with PCAÂ¶

```python
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA
wine_norm=scale(wine)
wine_norm
```

```python
pca=PCA()
pca_values=pca.fit_transform(wine_norm)
```

```python
pca_values
```

```python
var=pca.explained_variance_ratio_
var
```

```python
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
pca.components_
```

```python
plt.plot(var1,color="blue")
```

```python
pca_values[:,0:2]
```

```python
# plot between PCA1 and PCA2 
x = pca_values[:,0]
y = pca_values[:,1]
#z = pca_values[:2:3]
plt.scatter(x,y)
```

```python
pca_data=pd.DataFrame(columns=["X","Y"])
pca_data.X=x
pca_data.Y=y
```

```python
pca_data
```

# Clustering with PCA Data

```python
from sklearn.cluster import KMeans
model=KMeans(n_clusters=6).fit(pca_data)
pca_data.plot(x="X",y="Y",c=model.labels_,kind="scatter",s=50,cmap=plt.cm.coolwarm_r)
```

```python
import sklearn
sklearn.metrics.silhouette_score(pca_data,model.labels_)
```

# Clustering with Original Data

```python
fig=plt.figure(figsize=(8,4))
WCSS=[]
for i in range(1,11): 
    model=KMeans(n_clusters=i).fit(wine_norm)
    WCSS.append(model.inertia_)
    
plt.plot(range(1,11),WCSS)
plt.title("The Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```

# Conclusion and Insights

### 1. Key Findings and Insights

Data Quality : The wine dataset was clean, with no missing values or duplicates, and ready for advanced analysis.
Principal Component Analysis (PCA):
PCA successfully reduced the 13 original features into a smaller set of components. The first 3 principal components explained approximately 66.5% of the total variance, with PC1 alone capturing 36.2%. A scree plot confirmed that most information can be retained with just 2 or 3 components. The scatter plot of PC1 vs PC2 visually highlighted potential grouping/clustering of wine types, indicating PCA's effectiveness in pattern discovery. Clustering Potential:
Post-PCA, the data is suitable for clustering algorithms like KMeans, which can efficiently group data using the reduced dimensions for better accuracy and computational performance.

### 2. Practical Implications of Using PCA and Clustering
PCA (Principal Component Analysis) :
Simplifies complex datasets by reducing dimensionality, making them easier to visualize and interpret. Helps to remove noise and redundant features, improving the performance of machine learning models. Ideal for data compression and speeding up computations in large datasets. Crucial for visualizing high-dimensional data (e.g., plotting 2D projections of 13D data).
Clustering (e.g., KMeans) :
Enables unsupervised grouping of data, revealing hidden patterns or segments (like customer segmentation, wine types, etc.). More effective after PCA, as reduced features help clustering algorithms focus on the most informative data. Useful for market analysis, recommendation systems, fraud detection, and more.

### 3.Recommendations for Using PCA and Clustering
Use PCA when:
The dataset has many features (high dimensionality). We need to visualize complex data in 2D or 3D. We want to remove noise and focus on important features. We aim to improve model speed and performance.
Use Clustering (e.g., KMeans) when:
We want to find hidden groups or patterns in unlabeled data. We need to perform market segmentation or classification without labels. We plan to apply clustering after PCA for better accuracy and efficiency.
Combine PCA + Clustering when:
We want to enhance clustering performance on high-dimensional data. We need faster computation and more meaningful clusters. We aim to reduce redundancy and highlight the most informative features before grouping.
